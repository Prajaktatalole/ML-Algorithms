### K means clustering

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(2,10):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)   # inertia==wcss
```

```python
wcss
```

```python
lst=[i for i in range(2,10)]  # no of clusters
plt.figure()
plt.plot(lst,wcss,marker='o',mfc='red',linestyle='--')
plt.xlabel('No of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
###we will take n_clusters=4
kmeans=KMeans(n_clusters=4)
y_pred_clusters=kmeans.fit_predict(X)
```

```python
y_pred_clusters
```

```python
data['K cluster']=y_pred_clusters
```

```python
data
```

```python
data.drop(['SR_NO','Engg_College'],axis=1).groupby('K cluster').mean()
```
